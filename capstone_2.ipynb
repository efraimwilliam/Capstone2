### Uji Distribusi Goodness Of Fit

Kita akan melakukan pengecekan distribusi menggunakan  goodness_of_fit

```python
from scipy.stats import goodness_of_fit
import scipy

result = goodness_of_fit(
    dist = scipy.stats.norm, 
    data = df['trip_distance'],
)
print(result)
```

Setelah dilakukan uji distribusi data trip_distance yaitu Goodness of Fit<br>

Hasil Menunjukan bahwa P-Value jauh dibawah nilai Aplha yaitu 0.0001. Sehingga H0 ditolak dan H1 diterima.

Kesimpulannya dalah **data trip_distance memiliki distribusi yang <u>tidak normal**</u>. Jadi karena distribusi data tidak normal, kita akan menggunakan **metode uji non-parametrik**

Jika kita lihat data lain seperti tip_amount, passenger_count, dan total_amount menunjukan grafik yang tidak berdistribusi normal. Hal tersebut terjadi karena banyak outlier pada data tersebut.

### Outlier Data - Trip Distance

Sekarang kita akan melakukan cleansing data ringan untuk mengurangi adanya outlier. Data cleansing ini  bertujuan agar ketika kita melakukan uji non-parametrik, hasil nya tidak terlalu jauh dengan yang diharapkan
<br>

Kita akan mencoba untuk mengahpus outlier berdasarkan IQR

```python
df
```

```python
q1_tripdistance = df['trip_distance'].quantile(0.25)
q3_tripdistance = df['trip_distance'].quantile(0.75)

iqr= q3_tripdistance-q1_tripdistance

data_min = q1_tripdistance-(1.5*iqr)
data_max = q3_tripdistance+(1.5*iqr)
print(data_min, data_max)
```

```python
df = df[(df['trip_distance'] >= data_min) & (df['trip_distance'] <= data_max)]

df
```

```python
fig, ax = plt.subplots(
    ncols=2,
    nrows=1,
    figsize = (12,5)
)

sns.histplot(
    data=df['trip_distance'], 
    kde=True, 
    ax=ax[0]
)

sns.histplot(
    data=df['passenger_count'], 
    kde=True, 
    ax=ax[1]
    )

ax[0].set_title('Distribution of Trip Distance')
ax[0].set_xlabel('Trip Distance')
ax[0].set_ylabel('Count')

ax[1].set_title('Distribution of Passenger Count')
ax[1].set_xlabel('Passenger Count')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

Setelah dilakukan data cleansing menggunakan metode IQR, terlihat data **trip_distance** diatas tetap pada data yang tidak berdistribusi normal. Data tersebut adalah Right Skewed

Melihat dari jumlah data yang berbeda pada weekdays dan weekends, maka analisis ini dilakukan dengan tujuan agar kita mengetahui apakah ada perbedaan pada tarif taksi di weekdays dan weekends.

Yang perlu kita lakukan pertama adalah, kita harus mengetahui durasi perjalanan dari titip Pick Up hingga Drop Off pelanggan. Tujuan kita menghitung durasi adalah nantinya kita ingin mengetahui berapa fare_amount per kilometer nya. Maka kita membutuhkan data trip_duration 

```python
trip_duration = df_final['lpep_dropoff_datetime'] - df_final['lpep_pickup_datetime']

df_final['trip_duration'] = trip_duration.dt.total_seconds() / 3600

df_final['trip_duration'] = df_final['trip_duration']

df_final
```

Saya melakukan perhitungan trip_duration dengan format jam.

```python
df_final['trip_duration'].describe()
```

Setelah kita telah mendapatkan durasi dari tiap perjalanan. Trip duration dilkukan untuk melihat hasil analisis yang nanti akan kita gunakan

Langkah selanjutnya adalah, kita akan melihat fare_amount per Kilometer. Hal ini dilakukan dengan data yang sudah kita miliki yaitu fare_amount dan trip_distance. Tujuannya adalah kita ingin melihat berapa fare_amount yang dihasilkan per kilometernya sehingga kita bisa memandingkan fare amount per kilometer yang ada pada weekdays dan weekend

```python
df_final['fare_per_km'] = df_final.apply(
    lambda x: x['fare_amount'] / x['trip_distance'],
    axis=1
)

df_final
```

#### Uji Hipotesis : Perbedaan fare_per_km pada Weekends dan Weekdays

Tujuan dilakukan analisis perbedaan Fare per Kilometer pada Weekdays dan Weekends dilakukan untuk melihat struktur tarif yang sudah ditetapkan, apakah struktur tarif tersebut sudah optimal atau belum. Hal ini dilakukan untuk meningkatkan pendapatan dari perusahaan. Jika permintaan tinggi pada hari kerja, apakah tarif juga cenderung lebih tinggi? Apakah perbedaan jumlah trip akan mempengaruhi Fare per Kilometer

<br>Hipotesis Null (H0) :<br>
Tidak ada perbedaan antara jumlah **fare_per_km** pada hari kerja dengan akhir pekan (fare_per_km_weekdays = fare_per_km_weekdays)<br>
<br>Hipotesis Alternatif (H1) :<br>
Ada perbedaan antara jumlah tip **fare_per_km** pada hari kerja dengan akhir pekan (fare_per_km_weekdays != fare_per_km_weekdays)

```python
weekdays_fare = df_final[df_final['week_days_cat']=='weekday']['fare_per_km']
weekends_fare = df_final[df_final['week_days_cat']=='weekend']['fare_per_km']


print(f"Jumlah data hari kerja: {len(weekdays_fare)}")
print(f"Jumlah data akhir pekan: {len(weekends_fare)}")
```

```python
from scipy.stats import mannwhitneyu

result_manwhiteneyu = mannwhitneyu(weekdays_fare, weekends_fare, alternative='two-sided')

print(result_manwhiteneyu)
```

Dari hasil uji Mann White U, nilai P-value jauh dibawah nilai P-Value, maka artinya adalah H0 ditolak. Sehingga nilai H1 diterima

```python
fig, ax = plt.subplots(
    ncols=2,
    nrows=1,
    figsize = (14, 6)
)


sns.boxplot(
    data=df_final, 
    x='week_days_cat', 
    y='fare_per_km', 
    hue='week_days_cat',
    ax=ax[0]
)

plt.title("Boxplot Fare per KM: Weekdays vs Weekends")

# KDE plot
sns.kdeplot(
    data=df_final, 
    x='fare_per_km',
    hue='week_days_cat',
    multiple="stack",
    ax=ax[1]
)


plt.xlabel("fare_per_km")
plt.title("Distribution of Fare per KM")


plt.tight_layout()
plt.show()
```

##### Insight 

Dari uji statistik Mann White U menunjukan bahwa nilai P-value nya adalah 7.96512772043524e-126. Sehingga kita gagal menolak H0. 

Kesimpulan nya adalah <u>**terdapat perbedaan jumlah fare_per_km pada weekday dibandingkan dengan pada weeekend**</u>.

Grafik diatas menunjukan bahwa ada perbedaan pada nilai median weekdays yang sedikit lebih tinggi dibandingkan dengan weekends

### Analisis 2

5. Apakah ada perbedaan signifikan dalam peringkat **tip_amount** untuk perjalanan yang diambil pada jam sibuk pagi dibandingkan dengan jam sibuk sore hari, dari lokasi penjemputan yang sama (**PULocationID**)?<br>

Dikutip dari https://www.blacklane.com/en/blog/travel/rush-hour-in-nyc/#when-is-rush-hour-in-new-york

Jam sibuk pada New York Taxi trip ada pada

Jam Pagi : 06.00 AM- 10 AM
Jam Sore : 16.00 PM - 20 AM

<br>Hipotesis Null (H0) :<br>
Tidak ada korelasi antara **tip_amount** dengan jam sibuk pagi hari dan sore hari<br>
<br>Hipotesis Alternatif (H1) :<br>
Ada korelasi antara **tip_amount** dengan jam sibuk pagi hari dan sore hari

```python
df_final
```

```python
df_final['pickup_hour_cat'] = df_final['lpep_pickup_datetime'].dt.hour.apply(
    lambda x: 'morning_peak' if x >= 6 and x <= 10 else ('evening_peak' if x >= 16 and x <= 20 else 'off_peak')
)

df_final = df_final[df_final['pickup_hour_cat'].isin(['morning_peak', 'evening_peak'])].copy()

df_final['pickup_hour_cat'].value_counts()
```

```python
unique_pulocation = df_final.groupby(['PULocationID', 'pickup_hour_cat'])['tip_amount'].mean().reset_index()

unique_pulocation
```

```python
paired_pu_location = unique_pulocation.pivot_table(
    index='PULocationID',
    columns='pickup_hour_cat',
    values='tip_amount',
    aggfunc='mean'
)

paired_pu_location.dropna(inplace=True)

paired_pu_location
```

```python
from scipy.stats import wilcoxon

sumrank, p_value = wilcoxon(paired_pu_location['morning_peak'], paired_pu_location['evening_peak'])

print(sumrank, p_value)
```

```python
df_plot_wilcoxon = pd.DataFrame({
        'TripDistance': np.concatenate([paired_pu_location['morning_peak'], paired_pu_location['evening_peak']]),
        'PeakHourCategory': ['Morning Peak'] * len(paired_pu_location['morning_peak']) + ['Evening Peak'] * len(paired_pu_location['evening_peak'])
    })

plt.figure(figsize=(8, 6))
sns.boxplot(x='PeakHourCategory', y='TripDistance', data=df_plot_wilcoxon)
plt.title('Median Trip Distance: Morning Peak vs. Evening Peak')
plt.xlabel('Kategori Jam Sibuk')
plt.ylabel('Median Jarak Perjalanan (Mil)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Hasil uji hipotesis Wilcoxon menunjukan bahwa nilai P-Value adalah 0.0013321403034777062 yang mana di bawah di nilai alpha. Artinya gagal menolak H0 di tolak, dan H1 diterima

Artinya adalah Ada korelasi antara **tip_amount** dengan jam sibuk pagi hari dan sore hari

#### Kesimpulan & Rekomendasi : 

Dikarenakan adanya korelasi antara **tip_amount** dengan jam sibuk pagi hari dan sore hari dengan hasil jumlah tip biasanya lebih besar di sore hari, perusahaan bisa memberikan tip juga kepada pengemudi di pagi hari (morning_peak) sehingga, jumlah taxi yang ada pada pagi hari dan sore hari seiimbang. Jika tidak, maka ada kemungkinan para pengemudi enggan bekerja di pagi hari, mengingat jumlah tip yang dihasilkan tidak sebesar pada sore hari.
